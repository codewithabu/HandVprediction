# **Hill and Valley Prediction using Logistics Regression**


# **OBJECTIVE**


Hill and valley prediction model is used to accurately predict whether a particular location is a hill or a valley. This can be done by using a machine learning  logistic regression algorithm. The model is trained on a dataset of geographical features and their corresponding classifications, and then evaluated on a separate test dataset to measure its performance.

The ultimate goal of a hill and valley prediction model is to provide a useful tool for identifying hills and valleys in various geographic locations. This can have important applications in fields such as geology, agriculture, and urban planning. For example, geologists can use the model to map the topography of an area and identify potential geological features, such as fault lines or oil deposits. Farmers can use the model to determine the best locations for crops or livestock. Urban planners can use the model to identify areas that are prone to flooding or landslides.

The following are some of the specific objectives of a hill and valley prediction model:

1. To develop a model that can accurately predict whether a location is a hill or a valley.
2. To identify the most important geographical features that contribute to the classification of a location as a hill or a valley.
3. To evaluate the performance of the model on a separate test dataset.
4. To apply the model to real-world problems, such as mapping the topography of an area or identifying potential geological features.

# **Understanding about Data set**

Each record represents 100 points on a two-dimensional graph. When plotted in order (from 1 through 100) as the Y coordinate, the points will create either a Hill (a "bump" in the terrain) or a Valley (a "dip" in the terrain). See the original source for some examples of these graphs.     

**1-100**: Labeled "V##". Floating point values (numeric), the X-values.

**101**: Labeled "Class". Binary (0, 1) representing (valley, hill)

# **X Variables Standardized**

Standardization of datasets is a common requirement for many machine learning estimators implemented in scikit-learn; they might behave badly if the individual features do not more or less look like standard normally distributed data: Gaussian with zero mean and unit variance.

Next approach is go for MinMax Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

```python
X.shape
```

# **Train Test Split**

```python
 from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, stratify= y, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# **Model Train**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

# **Model Prediction**

```python
y_pred = lr.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

# **Probability of Each Predicted Class**

```python
lr.predict_proba(X_test)
```

# **Model Evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
